```python
############################################home work
elements = web.find_elements("xpath", "//div[@class='a-section a-spacing-base']")
for i in elements:
    print(i.text)
```

```python
###################################page 1
dic_of_out = []
id = 1

for element in elements:
    price_elements = element.find_elements("xpath", ".//span[@class='a-price-whole']")
    product_elements = element.find_elements("xpath", ".//h2[@class='a-size-base-plus a-spacing-none a-color-base a-text-normal']")
    image_element = element.find_elements("xpath", ".//img[@class='s-image']")
    rating_elements = element.find_elements("xpath", ".//i[contains(@class, 'a-icon-star-small')]//span[@class='a-icon-alt']")
    product_info = {
        'id': id,
        "product description": product_elements[0].text if product_elements else "No description",
        "rating": rating_elements[0].get_attribute("innerHTML") if rating_elements else "No rating",
        "price": price_elements[0].text if price_elements else "Price not found",
        "image": image_element[0].get_attribute("src") if image_element else "No image found"
    }
    dic_of_out.append(product_info)
    id += 1
```

```python
def page(url):
    dic_of_page2 = []
    id = 1
    web.get(url)
    elements = web.find_elements("xpath", "//div[@class='a-section a-spacing-base']")
    for element in elements:
        price_elements = element.find_elements("xpath", ".//span[@class='a-price-whole']")
        product_elements = element.find_elements("xpath", ".//h2[@class='a-size-base-plus a-spacing-none a-color-base a-text-normal']")
        image_element = element.find_elements("xpath", ".//img[@class='s-image']")
        rating_elements = element.find_elements("xpath", ".//i[contains(@class, 'a-icon-star-small')]//span[@class='a-icon-alt']")
        product_info = {
            'id': id,
            "product description": product_elements[0].text if product_elements else "No description",
            "rating": rating_elements[0].get_attribute("innerHTML") if rating_elements else "No rating",
            "price": price_elements[0].text if price_elements else "Price not found",
            "image": image_element[0].get_attribute("src") if image_element else "No image found"
        }
        dic_of_page2.append(product_info)
        id += 1
        return dic_of_page2
```
