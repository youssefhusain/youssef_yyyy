# Problem 1

Use numpy to simulate the vector outcomes for repeating the following random experiments $1000$ times and use the relative frequency to compute the probability of the corresponding events:
    
1. **Experiment:** Flipping a fair coin, **Event:** A head occurs.
     
Observe how changing the number of repetition affects the computed probabilities.

**Hints:** 
- Use [numpy.random.randint](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html) or [numpy.random.choice](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html) to simulate the experiment.
- Other useful functions [numpy.arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html), [numpy.cumsum](https://numpy.org/doc/stable/reference/generated/numpy.cumsum.html), [numpy.sum](https://numpy.org/doc/stable/reference/generated/numpy.sum.html), and [matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
trials = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]
for n in trials:
    results = np.random.choice([0, 1], size=n, p=[0.5, 0.5])
    probability = np.sum(results) / n

    print(f"n = {n}, P(e) = {probability}")
    print("-----------------------------------------")
```

```python
num_trials = 1000  
coin_outcomes = [0, 1]  # 0 for tails, 1 for heads
results = np.random.choice(coin_outcomes, size=num_trials, p=[0.5, 0.5])
cumulative_heads = np.cumsum(results)
relative_frequency = cumulative_heads / np.arange(1, num_trials + 1)
plt.plot(np.arange(1, num_trials + 1), relative_frequency, label="Relative Frequency of Heads")
plt.axhline(y=0.5, color='r', linestyle='--', label="Theoretical Probability (0.5)")
plt.xlabel("Number of Trials")
plt.ylabel("Probability of Heads")
plt.title("Convergence of Probability of Heads as Number of Trials Increases")
plt.legend()
plt.show()
final_probability = relative_frequency[-1]
print(f"Final probability of heads after {num_trials} trials: {final_probability:.4f}")
```

2. **Experiment:** Rolling a fair die, **Event:** Number 5 occurs.

```python
probabilities=[]
trials = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]
for n in trials:
    results = np.random.choice([1, 2, 3, 4, 5, 6], size=n, p=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
    probability = np.sum(results == 5) / n
    print(f"n = {n}, P(e) = {probability}")
    probabilities.append(probability)
    print("-----------------------------------------")
```

```python
plt.plot(trials, probabilities, marker='o', label="Probability of Rolling a 5")
plt.axhline(y=1/6, color='r', linestyle='--', label="Theoretical Probability (1/6)")
plt.xscale('log')  
plt.xlabel("Number of Trials (log scale)")
plt.ylabel("Probability of Rolling a 5")
plt.title("Convergence of Probability of Rolling a 5 as Number of Trials Increases")
plt.legend()
plt.grid(True)
plt.show()
```

```python
num_trials = 1000  
die_outcomes = [1, 2, 3, 4, 5, 6]  
results = np.random.choice(die_outcomes, size=num_trials, p=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
cumulative_fives = np.cumsum(results == 5)
relative_frequency = cumulative_fives / np.arange(1, num_trials + 1)
plt.plot(np.arange(1, num_trials + 1), relative_frequency, label="Relative Frequency of Rolling a 5")
plt.axhline(y=1/6, color='r', linestyle='--', label="Theoretical Probability (1/6)")
plt.xlabel("Number of Trials")
plt.ylabel("Probability of Rolling a 5")
plt.title("Convergence of Probability of Rolling a 5 as Number of Trials Increases")
plt.legend()
plt.show()
final_probability = relative_frequency[-1]
print(f"Final probability of rolling a 5 after {num_trials} trials: {final_probability:.4f}")
```

# Problem 2

Simulate the experiment of tossing $1000$ times a coin with probability of **head (coded as $1$) equal to $0.4$** and probability of **tail (coded as $0$) equal to $0.6$**. How can you check if the simulation of the experiment of been running as expected.

**Hint**: Use [matplotlib.pyplot.hist¶
](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

```python
num_trials = 1000  
coin_outcomes = [0, 1]  
probabilities = [0.6, 0.4]  
results = np.random.choice(coin_outcomes, size=num_trials, p=probabilities)
relative_frequency_heads = np.sum(results == 1) / num_trials
relative_frequency_tails = np.sum(results == 0) / num_trials
print(f"Relative frequency of heads: {relative_frequency_heads:.4f}")
print(f"Relative frequency of tails: {relative_frequency_tails:.4f}")
plt.hist(results, bins=[-0.5, 0.5, 1.5], edgecolor='black', alpha=0.7, rwidth=0.8)
plt.xticks([0, 1], ['Tails (0)', 'Heads (1)'])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Histogram of Coin Toss Outcomes")
plt.show()
```

# Problem 3: Independence

Use NumPy to create simulations and compute **empirical  probability** for the following outcomes. Compare the results with the **theoretical probability**.

1. Two fair coin flips produce exactly two heads

```python
# Parameters
num_tests = 1_000_000# simulate 1 million tests of two fair coin flips
coin_outcomes = [0, 1]  
flip1 = np.random.choice(coin_outcomes, size=num_tests, p=[0.5, 0.5])
flip2 = np.random.choice(coin_outcomes, size=num_tests, p=[0.5, 0.5])
two_heads = (flip1 == 1) & (flip2 == 1)
empirical_probability = np.sum(two_heads) / num_tests
print(f"Empirical probability of exactly two heads: {empirical_probability:.6f}")
print(f"Theoretical probability of exactly two heads: 0.25")
```

2. Three fair coin flips produce exactly one head

```python
num_tests = 1_000_000  
coin_outcomes = [0, 1]  
flip1 = np.random.choice(coin_outcomes, size=num_tests, p=[0.5, 0.5])
flip2 = np.random.choice(coin_outcomes, size=num_tests, p=[0.5, 0.5])
flip3 = np.random.choice(coin_outcomes, size=num_tests, p=[0.5, 0.5])
exactly_one_head = (flip1 + flip2 + flip3) == 1
empirical_probability = np.sum(exactly_one_head) / num_tests
print(f"Empirical probability of exactly one head: {empirical_probability:.6f}")
print(f"Theoretical probability of exactly one head: 0.375")
```

3. Three biased coin flips with P(H) = 0.6 produce exactly one head

```python
num_tests = 1_000_000  # Number of tests
coin_outcomes = [0, 1]  
p_head = 0.6  
p_tail = 1 - p_head 
flip1 = np.random.choice(coin_outcomes, size=num_tests, p=[p_tail, p_head])
flip2 = np.random.choice(coin_outcomes, size=num_tests, p=[p_tail, p_head])
flip3 = np.random.choice(coin_outcomes, size=num_tests, p=[p_tail, p_head])
exactly_one_head = (flip1 + flip2 + flip3) == 1
empirical_probability = np.sum(exactly_one_head) / num_tests
print(f"Empirical probability of exactly one head: {empirical_probability:.6f}")
print(f"Theoretical probability of exactly one head: 0.288")
```

# Problem 4:Random Variables Statistical Tools
Simulate Rolling a fair die $1000$ times. Then calculate the following:

- Sample mean
- Sample variance
- Sample standard deviation

Compare the results with the theoretical probability.

```python
num_trials = 1000  # Number of die rolls
die_outcomes = [1, 2, 3, 4, 5, 6]  
results = np.random.choice(die_outcomes, size=num_trials)
sample_mean = np.mean(results)
sample_variance = np.var(results, ddof=0)  
sample_std_dev = np.std(results, ddof=0) 
print(f"Sample mean: {sample_mean:.3f}")
print(f"Sample variance: {sample_variance:.6f}")
print(f"Sample standard deviation: {sample_std_dev:.6f}")
print(sample_mean)
print(sample_variance)
print(sample_std_dev)
```

## Temperature and Apparent Temperature

```python
df = pd.read_csv('weatherHistory.csv')
temperature = df['Temperature (C)'].to_numpy().reshape(-1, 1)
apparent_temp = df['Apparent Temperature (C)'].to_numpy().reshape(-1, 1)
humidity = df['Humidity'].to_numpy().reshape(-1, 1)
cov_temp_apparent = np.cov(temperature.T, apparent_temp.T)
corr_temp_apparent = np.corrcoef(temperature.T, apparent_temp.T)
cov_temp_humidity = np.cov(temperature.T, humidity.T)
corr_temp_humidity = np.corrcoef(temperature.T, humidity.T)
print("## Temperature and Apparent Temperature")
print("Covariance: ", cov_temp_apparent[0, 1])
print("---------------------------------------------------")
print("np.cov(): \n", cov_temp_apparent)
print("---------------------------------------------------")
print("Correlation Coeff: ", corr_temp_apparent[0, 1])
print("---------------------------------------------------")
print("np.corrcoef(): \n", corr_temp_apparent)
print("\n## Temperature and Humidity")
print("Covariance: ", cov_temp_humidity[0, 1])
print("---------------------------------------------------")
print("np.cov(): \n", cov_temp_humidity)
print("---------------------------------------------------")
print("Correlation Coeff: ", corr_temp_humidity[0, 1])
print("---------------------------------------------------")
print("np.corrcoef(): \n", corr_temp_humidity)
```

## Temperature and Humidity

```python
temperature = df['Temperature (C)'].to_numpy().reshape(-1, 1)
humidity = df['Humidity'].to_numpy().reshape(-1, 1)
cov_temp_humidity = np.cov(temperature.T, humidity.T)
corr_temp_humidity = np.corrcoef(temperature.T, humidity.T)
print("Covariance: ", cov_temp_humidity[0, 1])
print("---------------------------------------------------")
print("np.cov(): \n", cov_temp_humidity)
print("---------------------------------------------------")
print("Correlation Coeff: ", corr_temp_humidity[0, 1])
print("---------------------------------------------------")
print("np.corrcoef(): \n", corr_temp_humidity)
```
